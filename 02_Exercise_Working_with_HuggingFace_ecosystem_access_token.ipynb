# Load a Pre-Trained BERT Model: bert-base-uncased

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
```

```python
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
```

```python
model = AutoModelForSequenceClassification.from_pretrained("bert-base-uncased")
```

# Log into Hugging Face Hub

## First, import notebook_login from huggingface_hub

```python
from huggingface_hub import notebook_login
```

## Next run notebook_login()
Copy and paste the access boken from your account in Hugging Face hub

```python
notebook_login()
```

# Save and Share a Model to the Hub with push_to_hub()

```python
model.push_to_hub("fine_tuned_10012023")
```

# Save a Model and Tokenizer to a Local Location

```python
saved_path = "./my_model_saved"
```

```python
model.save_pretrained(saved_path)
```

```python
tokenizer.save_pretrained(saved_path)
```

# Load the Saved Model and Tokenizer from a Local Location

```python
model_loaded_from_local = AutoModelForSequenceClassification.from_pretrained(saved_path)
```

```python
tokenizer_loaded_from_local = AutoTokenizer.from_pretrained(saved_path)
```
